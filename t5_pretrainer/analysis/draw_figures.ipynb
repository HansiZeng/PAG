```python
import numpy as np
import matplotlib.pyplot as plt

X1 = np.linspace(0,5,5)
X2 = np.linspace(1,6,5)
Xs = [X1,X2]
Ys = [X1**2,X2**2]
Zs = [X1**3,X2**3]

# Marker size
size = 100

fig, axs = plt.subplots(1,2)
for x,y,z, in zip(Xs,Ys,Zs):
    axs[0].scatter(x,y,linewidth=1.5,s=size,facecolors='#0571b0',marker='o',alpha=0.5,label='A label')
    axs[1].scatter(x,z,linewidth=1.5,s=size,facecolors='#92c5de',marker='o',alpha=0.5,label='Another label')

plt.figlegend(loc = 'upper right', ncol=1, labelspacing=0.)
plt.show()
```
